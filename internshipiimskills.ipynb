```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('/content/Traffic_data 1.csv', encoding='latin1')

# Select only numerical columns for clustering
x = data.select_dtypes(include=np.number).dropna() # Redefine x

# Refit the KMeans model on the first two columns for visualization
algorithm = (KMeans(n_clusters= 3, init='k-means++', n_init=10, max_iter=300,
                    tol=0.0001,  random_state= 111  , algorithm='elkan'))
algorithm.fit(x.iloc[:, :2]) # Fit on the first two columns
labelsl = algorithm.labels_ # Get labels after fitting on 2 columns
centroids = algorithm.cluster_centers_ # Get centroids after fitting on 2 columns

h = 0.02
x_min, x_max = x.iloc[:, 0].min() - 1, x.iloc[:,0].max()+1
y_min, y_max = x.iloc[:,1].min() - 1, x.iloc[:,1].max()+1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
z = algorithm.predict(np.c_[xx.ravel(), yy.ravel()])

# Reshape the predictions
Z = z.reshape(xx.shape)

# Plot the decision boundary
plt.figure(1 , figsize = (15 , 10))
plt.clf()
plt.imshow(Z, interpolation='nearest', extent=(xx.min(), xx.max(), yy.min(), yy.max()), cmap = plt.cm.Pastel1, aspect = 'auto', origin='lower')

# Plot the data points
plt.scatter(x.iloc[:, 0], x.iloc[:, 1], c=labelsl, s=50, cmap='viridis')

# Plot the centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='red', edgecolors='black', label='Centroids')

plt.title('KMeans Clustering with Decision Boundaries')
plt.xlabel(x.columns[0])
plt.ylabel(x.columns[1])
plt.legend()
plt.show()
```
